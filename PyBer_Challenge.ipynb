### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city typeur
count_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
count_rides_by_type
```

```python
#total_urban_drivers = pyber_data_df.loc[pyber_data_df["type"]== "Urban"].sum()["driver_count"]
#total_urban_drivers
total_urban_drivers = pyber_data_df.loc[pyber_data_df["type"]== "Urban", ["driver_count"]].sum()
total_urban_drivers
```

```python
# 2. Get the total drivers for each city type
#total_drivers_by_type = pyber_data_df.groupby(["type"])["driver_count"].sum()
#total_drivers_by_type
#total_drivers_by_type = pyber_data_df.groupby(["type"]).sum(["driver_count"])
#total_drivers_by_type
total_drivers_per_city = city_data_df.groupby(["type"])["driver_count"].sum()
total_drivers_per_city
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fares_by_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fares_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
#avg_fares_per_driver = pyber_data_df.groupby(["type"])
avg_fares_per_driver = sum_fares_by_type / total_drivers_per_city
avg_fares_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
#pyber_summary_df = pd.DataFrame({'Total Rides': [count_rides_by_type],
                                # 'Total Drivers': [total_drivers_per_city],
                                #'Total Fares': [sum_fares_by_type],
                                #'Average Fare Per Ride': [avg_fares_by_type],
                                #'Avg Fare Per Driver':[avg_fares_per_driver]})
#pyber_summary_df
#type_test = avg_fares_per_driver[["Type", "Average"]]
#type_test = type_test.set_index("Type")
#type_test.head()
pyber_summary_df = pd.DataFrame(
          {"Total Rides": count_rides_by_type.map("{:,}".format),
          "Total Drivers": total_drivers_per_city.map("{:,}".format),
          "Total Fares": sum_fares_by_type.map("${:,.2f}".format),
          "Average Fare per Ride": avg_fares_by_type.map("${:,.2f}".format),
          "Average Fare per Driver": avg_fares_per_driver.map("${:,.2f}".format)})


pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

sum_fares = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_fares
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_fares = sum_fares.reset_index()
sum_fares
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_types_pivot = sum_fares.pivot(index="date", columns="type", values="fare")
sum_types_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# Create a new DataFrame from the pivot table DataFrame by filtering for the given dates, '2015-01-01':'2015-12-31'. 
filtered_dates_pivot_df = sum_types_pivot.loc['2019-01-01':'2019-04-29']
filtered_dates_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Stepsum 8.
filtered_dates_pivot_df.index = pd.to_datetime(filtered_dates_pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
filtered_dates_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_sum_fares_df = filtered_dates_pivot_df.resample('W').sum()
weekly_sum_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# Plot the resampled DataFrame.

weekly_rides = weekly_sum_fares_df.plot(figsize = (20,6))
# Add a title 
weekly_rides.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
weekly_rides.set_xlabel("Month")
weekly_rides.set_ylabel("Fare($USD)")


plt.savefig('pyber_fare_summary.png')
plt.tight_layout()
```

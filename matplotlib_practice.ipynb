```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create the labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the limit between 0 and 45
plt.ylim(0, 45)
# Create the title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label= 'City of Boston')
# add x-label - date
ax.set_xlabel('Date')
# add y-label - fare($)
ax.set_ylabel("Fare($)")
# add y-axis limit -45
ax.set_ylim(10, 40)
# Title
ax.set_title("PyBer Fare by Month")
#add grid lines
ax.grid()
```
